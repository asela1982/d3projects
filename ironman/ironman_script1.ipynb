```python
# import relevant libraries
import pandas as pd
import numpy as np
import time
```

### extract main table

```python
# instantiate an empty list to store the dataframes
appended_data = []
# loop through each page to extract tabular data
for i in range(1,47):
    #  make the script sleep for few seconds to factor any delays on page loading
    time.sleep(5)
    #  construct the url
    url =  "https://www.multisportaustralia.com.au/races/15238/events/1?page="+str(i)
    #  convert the tablular data into a pandas dataframe    
    df = pd.read_html(url)[0]
    # add the dataframes to the list
    appended_data.append(df)
# append the dataframes using pandas concat method
result = pd.concat(appended_data,axis=0)
```

```python
result['Bib'] = result['Name'].str.extract(r'\(([^)]+)\)')
result['Category_pos'] = result['Category (POS)'].str.extract(r'\(([^)]+)\)')
result['Gender_pos'] = result['Gender (POS)'].str.extract(r'\(([^)]+)\)')
result['Name'] = result['Name'].str.replace(r'\(([^)]+)\)',"")
result['Category'] = result['Category (POS)'].str.replace(r'\(([^)]+)\)',"")
result['Gender'] = result['Gender (POS)'].str.replace(r'\(([^)]+)\)',"")
```

```python
result
```

```python
result['Category'] = result['Category'].str.strip()
result['Gender'] = result['Gender'].str.strip()
```

```python
result
```

```python
result.to_csv('data/results.csv',index=False)
```

### extract individual performances

```python
url_ind = "https://www.multisportaustralia.com.au/races/15238/events/1/results/individuals/1?page=1"
```

```python
swim = pd.read_html(url_ind)[0]
```

```python
t1 = pd.read_html(url_ind)[1]
```

```python
bike = pd.read_html(url_ind)[2]
```

```python
t2 = pd.read_html(url_ind)[3]
```

```python
run = pd.read_html(url_ind)[4]
```

```python
main = pd.concat([swim,t1,bike,t2,run], ignore_index=True)
```

```python
main.head()
```

```python
# extract Speed(kms per hour) from the Speed/Pace colum
def extract_speed(x):
    return x.split('/')[0]
```

```python
# extract Pace(time for a km) from the Speed/Pace colum
def extract_pace(x):
    return x.split('/')[1]
```

```python
main['Speed'] = main['Speed/Pace'].apply(extract_speed)
```

```python
main['Pace'] = main['Speed/Pace'].apply(extract_pace)
```

```python
main.head()
```

```python
main[['Location','Split','Pos','Div','Gen','Speed','Pace']].to_csv("data/view2.csv",index=False)
```
